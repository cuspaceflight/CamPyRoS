```python
from scipy.io import netcdf
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
forcast_location="data/wind/historic/radar-mst_capel-dewi_20200101_st300_cartesian_v3.nc"
#http://data.ceda.ac.uk/badc/mst/data/mst-products-v3/st-mode/cartesian/2020/01
f = netcdf.NetCDFFile(forcast_location,'r')
data=pd.DataFrame({"altitude":f.variables["altitude"][:]*1,"eastward_wind":f.variables["eastward_wind"][:][0],"northward_wind":f.variables["northward_wind"][:][0]})
```

```python
data["wind_speed"]=(data["eastward_wind"]**2+data["northward_wind"]**2)**.5
```

```python
data.plot(x="altitude",y="wind_speed")
```
