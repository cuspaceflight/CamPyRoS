```python
%matplotlib notebook
import campyros.statistical as stats 
import campyros.plot as plt
import matplotlib.pyplot as pyplot
#pyplot.rcParams["figure.figsize"] = (15,10)#Comment this out if using matplotlib notebook
```

## Run the stats model
This will take a long time, if you're just testing reduce the number of runs to around 100 in the json file.
Approximatly 5000 runs are required for the results to be normally distributed for a full model.

```python
model = stats.StatisticalModel("stats_settings.json")
model.run_model()
```

## Loads results and calculate stats

```python
save_loc = "results/stats_example/stats_example"
max_itts = 2500

l_mu,l_cov,ap_mu,ap_cov,ap_dat,l_dat,x,y,z,t=stats.analyse(save_loc,max_itts)
```

Plot of landing sites

```python
plt.stats_landing(l_mu,l_cov,data=l_dat)
```

Plot of apogees

```python
plt.stats_apogee(ap_mu,ap_cov,apogee=ap_dat,
                 landing_mu=l_mu,landing_cov=l_cov,landing=l_dat)
```

Plot of trajectories

```python
plt.stats_trajectories(x,y,z)
```

Plot of trajectories with landing and apogee covariant elipses

```python
plt.stats_trajectories(x,y,z,
                       apogee_mu=ap_mu,apogee_cov=ap_cov,
                       landing_mu=l_mu,landing_cov=l_cov)
```

Plot of altitude time with standard errors
Warning, the following graph takes a long time to run

```python
plt.stats_alt(z,t,show_means=True)
```

I think the means/error bounds are wiggly because the timesteps get long on the way down.
